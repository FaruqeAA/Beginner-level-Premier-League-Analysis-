```python
import pandas as pd

df = pd.read_csv("E0.csv")
print(df.shape)
print(df.head())
```

```python
import matplotlib.pyplot as plt

home-goals = df.groupby("HomeTeam")["FTHG"].sum()
away-goals = df.groupby("AwayTeam")["FTAG"].sum()
total_goals = (home-goals + away-goals).sort_values(ascending=False)

plt.figure(figsize=(14, 6))
plt.bar(total_goals.index, total_goals.values, color="royalblue")
plt.title("Total Goals Scored by Each Team — Premier League 2023/24")
plt.xlabel("Team")
plt.ylabel("Goals")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python

homewins = df[df["FTR"] == "H"]["HomeTeam"].value_counts()
awaywins = df[df["FTR"] == "A"]["AwayTeam"].value_counts()

import pandas as pd
results_df = pd.DataFrame({"Home Wins": homewins, "Away Wins": awaywins}).fillna(0)
results_df = results_df.sort_values("Home Wins", ascending=False)

results_df.plot(kind="bar", figsize=(14, 6), color=["royalblue", "tomato"])
plt.title("Home vs Away Wins by Team — Premier League 2023/24")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
homedraws = df[df["FTR"] == "D"]["HomeTeam"].value_counts()
awaydraws = df[df["FTR"] == "D"]["AwayTeam"].value_counts()
total_draws = (homedraws + awaydraws).sort_values(ascending=False)

plt.figure(figsize=(14, 6))
plt.bar(total_draws.index, total_draws.values, color="mediumseagreen")
plt.title("Most Draws by Team — Premier League 2023/24")
plt.xlabel("Team")
plt.ylabel("Draws")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
homeconceded = df.groupby("HomeTeam")["FTAG"].sum()
awayconceded = df.groupby("AwayTeam")["FTHG"].sum()
total_conceded = (homeconceded + awayconceded).sort_values(ascending=True)

plt.figure(figsize=(14, 6))
plt.bar(total_conceded.index, total_conceded.values, color="crimson")
plt.title("Goals Conceded by Each Team — Premier League 2023/24")
plt.xlabel("Team")
plt.ylabel("Goals Conceded")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
total_goals = df["FTHG"].sum() + df["FTAG"].sum()
avg_goals_per_game = total_goals / len(df)
highest_home_score = df["FTHG"].max()
highest_away_score = df["FTAG"].max()

df["TotalGoals"] = df["FTHG"] + df["FTAG"]
highest_scoring_match = df.loc[df["TotalGoals"].idxmax()]

print("=" * 35)
print("   PREMIER LEAGUE 2023/24 — KEY STATS")
print("=" * 35)
print(f"Total goals scored:        {total_goals}")
print(f"Average goals per game:    {avg_goals_per_game:.2f}")
print(f"Most goals by a home team: {highest_home_score}")
print(f"Most goals by an away team:{highest_away_score}")
print(f"\nHighest scoring match:")
print(f"  {highest_scoring_match['HomeTeam']} {int(highest_scoring_match['FTHG'])} - {int(highest_scoring_match['FTAG'])} {highest_scoring_match['AwayTeam']}")
print(f"  Date: {highest_scoring_match['Date']}")
print("=" * 35)
```
